**About the 2024/25 La Liga Season**

The 2024/25 La Liga season, officially known as La Liga EA Sports due to sponsorship, was the 94th edition of Spain’s top-flight football competition. The season ran from August 15, 2024, to May 25, 2025.

The league featured 20 clubs from across Spain, competing in a traditional double round-robin format — each team faced every other team twice, once at home and once away, totaling 38 matchdays. The fixture schedule was determined by a draw before the season began. As per standard rules, teams earned 3 points for a win, 1 point for a draw, and none for a loss. Final standings were based on the total number of points accumulated throughout the season.

Barcelona secured their 28th league title on May 15, 2025, with two games remaining, following a 2–0 victory over city rivals Espanyol.

#**Player-Level Analysis**

## **1- Distribution of Players by Nationality**

```python
# Counting the number of players per nation
nation_counts=laliga['Nation'].value_counts()

# Group nations with 5 or fewer players into a single 'Others' category
threshold=5
nation_filtered=nation_counts.copy()
others_sum=nation_filtered[nation_filtered<=threshold].sum()
nation_filtered=nation_filtered[nation_filtered>threshold]
nation_filtered['Others']=others_sum
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
ax=sns.barplot(x=nation_filtered.index, y=nation_filtered.values, palette="viridis", hue=nation_filtered.index, legend=False)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2,
            p.get_height() + 0.5,
            int(p.get_height()),
            ha='center',
            fontsize=10,
            color='black')

plt.title('Number of Players by Nation (5 or fewer grouped as Others)', fontsize=14, pad=12)
plt.xlabel('Nation')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This chart shows how many players in La Liga 2024/25 come from each country. Not surprisingly, most of the players are from Spain. Other countries like Argentina, France, and Brazil also have a decent number of players. To keep the chart simple, countries with 5 or fewer players are grouped under “Others.”

## **5- Standout Players Radar Profile (Multidimensional)**

```python
# Filtering standout players
standouts=laliga[
    (laliga['Gls/90'] > 0.2) &
    (laliga['Ast/90'] > 0.1) &
    (laliga['xG/90'] > 0.1) &
    (laliga['xAG/90'] > 0.1)
].copy()

# Normalizing metrics using global max
metrics=['Gls/90', 'Ast/90', 'xG/90', 'xAG/90']
for metric in metrics:
    global_max=laliga[metric].max()
    standouts[metric]=standouts[metric]/global_max

# Calculating overall performance score
standouts['AvgScore']=standouts[metrics].mean(axis=1)

# Selecting top 3 players based on average score
players=standouts.sort_values(by='AvgScore', ascending=False).head(3)

# Radar plot
labels=metrics
num_vars=len(labels)
angles=[n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles+=angles[:1]

plt.figure(figsize=(7, 7))
for i, row in players.iterrows():
    values=row[metrics].tolist()
    values+=values[:1]
    plt.polar(angles, values, label=row['Player'])

plt.xticks(angles[:-1], labels)
plt.title('Top Multifaceted Players (Avg. Normalized Score)',fontsize=14, pad=35)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))
plt.grid(True)
plt.tight_layout()
plt.show()
```

This chart highlights the top multifaceted players in La Liga based on their average normalized scores across key performance metrics. The players featured —**Ferrán Torres**, **Pau Victor**, and **Raphinha**— are evaluated on their contributions per 90 minutes in areas such as expected goals (xG/90), expected assisted goals (xAG/90), and assists (Ast/90).

#**Team-Level & Correlation Analysis**

## **6- Average Age of the Teams**

```python
# Calculating average age per team & overall average age
team_age_avg=laliga.groupby('Team')['Age'].mean().sort_values()
overall_avg=laliga['Age'].mean()

# Generating color palette based on number of teams
colors=sns.color_palette('coolwarm', len(team_age_avg))

# Creating the plot
plt.figure(figsize=(10, 6))
sns.barplot(x=team_age_avg.values, y=team_age_avg.index, hue=team_age_avg.index, palette='coolwarm', legend=False)

# Vertical line showing the league-wide average age
plt.axvline(overall_avg, color='black', linestyle='--', linewidth=1.5, label=f'League Avg ({overall_avg:.1f})')

# Title & axis labels and plot
plt.title('Average Age of Teams in La Liga (2024-25)', fontsize=14, pad=12)
plt.xlabel('Average Age')
plt.ylabel('Team')
plt.xlim(20, 30)
plt.legend()
plt.tight_layout()
plt.show()
```

This chart displays the average age of squads for each team in La Liga during the 2024-25 season. The teams are listed alongside horizontal bars representing their average age, with the league-wide average calculated at 25.4 years. Barcelona, Valencia, and Real Sociedad appear among the younger squads, while teams like Atlético Madrid and Rayo Vallecano anchor the older end of the spectrum.

## **7- Comparison of Expected (xG) vs Actual Total Goals by Team**

```python
# Grouping by team and calculate total xG and Goals
team_sum = laliga.groupby('Team').agg({
    'xG': 'sum',
    'Goals': 'sum'
}).reset_index()

# Creating an interactive scatter plot
fig = px.scatter(
    team_sum,
    x='xG',
    y='Goals',
    text='Team',
    color='Team',
    title='Expected (xG) vs Actual Total Goals (2024-25)',
    labels={'xG': 'Total Expected Goals (xG)', 'Goals': 'Total Goals'},
    width=900,
    height=600
)

# Adjusting marker size and layout for clarity
fig.update_traces(marker=dict(size=15), textposition='top center')
fig.update_layout(showlegend=False)  # Optional: hides side legend to reduce clutter

# Showing the figure
fig.show()
```

This scatter plot visualizes the relationship between total expected goals (xG) and actual goals scored by La Liga teams during the 2024-25 season. The x-axis represents expected goals, while the y-axis shows the actual number of goals scored. **Barcelona** clearly leads in both categories, indicating strong finishing ability.

## **8- Relationship Between Cards and Team Performance**

```python
# Grouping by team and summing yellow cards, red cards, and xG
team_cards=laliga.groupby('Team').agg({
    'Yellow_Cards': 'sum',
    'Red_Cards': 'sum',
    'xG': 'sum'
}).reset_index()

# Calculating total cards
team_cards['Total Cards']=team_cards['Yellow_Cards'] + team_cards['Red_Cards']

# Regression plot with 95% confidence interval
plt.figure(figsize=(10, 6))
sns.regplot(
    data=team_cards,
    x='Total Cards',
    y='xG',
    ci=95,
    scatter_kws={'s': 70, 'color': 'crimson'},
    line_kws={'color': 'darkblue'}
)

# Annotating each data point with the team name
for i in range(team_cards.shape[0]):
    plt.text(team_cards['Total Cards'][i] + 0.3,
             team_cards['xG'][i],
             team_cards['Team'][i],
             fontsize=9)

# Plot
plt.title('Relationship Between Cards and Team Expected Goals (xG)', fontsize=14)
plt.xlabel('Total Cards (Yellow + Red)')
plt.ylabel('Total Expected Goals (xG)')

plt.grid(True)

plt.tight_layout()
plt.show()
```

This analysis reveals an inverse relationship between cards and xG in La Liga 2024-25. Top teams -**Barcelona**, **Real Madrid**- with high xG typically receive fewer cards due to controlled play, while weaker teams -**Valladolid**, **Las Palmas**- with low xG commit more fouls, leading to more cards. The data suggests defensive fragility (high cards) correlates with poor attack (low xG), but causation requires deeper study.

## **9- Correlation Analysis of Player Performance Metrics per 90 Minutes**

```python
# Selecting only the columns that represent 'per 90 minutes'
per90_columns=[col for col in laliga.columns if '/90' in col]

# Correlation matrix between these metrics
corr_matrix=laliga[per90_columns].corr()

# Creating heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0,
            linewidths=0.5, square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Between Per 90 Stats in La Liga 2024/25', fontsize=14, pad=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

*   There's a strong correlation (0.64) between goal involvement (**GIs/90**) and
expected goals (**xG/90**), showing that players who get into scoring positions contribute more to goals.
*   Assists (**Ast/90**) and expected assists (**xAG/90**) are very strongly linked (0.75), meaning created chances often result in actual assists.
*   The low correlations between **Ast/90 and GIs/90**, as well as **xG/90 and xAG/90**, suggest that players tend to be either goal-scorers or playmakers, but not both.
